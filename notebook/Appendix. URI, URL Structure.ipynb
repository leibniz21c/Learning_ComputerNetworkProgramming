Appendix.URI, URL Structure
=======
---
### 1. URI(Uniform Resource Identifier)
### 2. URL Structure(Uniform Resource Locator)
=======
#### 1. URI(Uniform Resource Identifier)
*Definition.URI*
URI이란, 웹 서버가 리소스를 고유하게 식별할 수 있도록 하는 것으로써, URL과 URN 두 가지가 있는데 일반적으로 URL을 사용한다.
- URL은 특정 서버의 한 리소스에 대해 구체적인 위치를 서술한다.
- URN은 리소스가 어디에 위치해 있든 찾을 수 있는 방식이다.

#### 2. URL Structure(Uniform Resource Locator)
2.1. URL 구조
URL은 scheme에 따라 문법이 모두 다르지만, 아래의 구조를 기반으로 선택적으로 사용.
```
<scheme>://<username>:<password>@<host>:<port>/<path>?<질의>#<프레그먼트>
```
1. Scheme(스킴)
	- 사용할 프로토콜을 말하며, 리소스에 어떻게 요청, 접근할 것인지를 명시한다.
	- 웹에서 주로 HTTP 프로토콜을 사용한다.
	- 종류 : ftp, mailto, rtsp 등
2. username and password
	- 어떤 서버들은 자신이 가지고 있는 데이터에 접근하기 위해서 사용자의 이름과 비밀번호를 요구한다
		- ex) ftp://victolee:12345@호스트/asd.xls
	- 만약 웹 서버가 사용자이름과 비밀번호를 요구하는 URL 스킹을 사용함에도 클라이언트가 이를 명시하지 않고 접근한다면, 기본적으로 사용자 이름 : 'anonumous', 비밀번호는 브라우저에서 제공하는 기본 값을 따른다. 
3. host and port
	- 하나의 host(컴퓨터)에는 여러 개의 Process(프로그램)이 각각의 socket을 사용하여 데이터 통신을 하고 있기 때문에, 각각의 소켓을 구분할 필요가 있다.
	- 이 때 소켓을 구분하는 역활이 Port이다.
	- 로컬에서 개발을 했을 때 접근하는 URL은 localhost:8080 등 이 있다.
	- 이처럼 서버에는 포트에 따라 소켓이 연결되어 있고, 포트 번호에 따라 다른 프로토콜이 사용될 수 있다.
	- HTTP 프로토콜에서 포트 번호를 명시하지 않으면, 80번 포트를 기본 값으로 사용.(Well-Known port), ex) http://www.google.com:80

4. path
	- 호스트에서 제공하는 자원의 경로를 의미
	- ex) https://movie.naver.com/**movie/running/current.nhn**

5. 질의
	- Query String(쿼리 스트링)이라고도 한다.
	- 클라이언트가 자원을 Get방식으로 요청할 때, 필요한 데이터를 함께 넘겨 줄 목적으로 사용
	- 개발할 때 함수를 호출하면 파라미터를 던져주는데, 이와 비슷
	- ex) http://localhost:3000/index?**id=3&page=1**

6. 프레그먼트
 - HTML에는 각각의 요소에 id 속성을 부여할 수 있다. URL에 프래그먼트를 전달하면 페이지가 해당 id가 있는 곳으로 스크롤이 이동한다.
- 여기서도 URL에 프래그먼트를 추가하면, 가장 마지막으로 이동한다.
- ex) [http://localhost:8888/notebooks/Desktop/Learning_ComputerNetworkProgramming/notebook/Appendix.%20URI%2C%20URL%20Structure.ipynb](http://localhost:8888/notebooks/Desktop/Learning_ComputerNetworkProgramming/notebook/Appendix.%20URI%2C%20URL%20Structure.ipynb/287#bottom)
 -> 주피터에선 안되는데 왠지 모르겠음.

2.2. URL 허용 문자
안전한 전송이란 클라이언트의 요청 URL 문자열이 손실되지 않고 서버측으로 전송되는 것이다.

안전한 전송을 위해 초기 URL 설계자들은 알파벳(ASCII)만으로 URL을 작성하도록 권장했음. 그런데 웹이 커져서 비영어권 나라에서 URL에 다른것도 사용. 

이에 ASCII 외의 문자들은 이스케이프 처리를 하여, ASCII 문자가 아닌 (완전하지 않은) 문자를 안전하도록 인코딩함.

인코딩을 통해 안전하지 않은 문자를 %기호로 시작하고  ASCII 코드로 표현되는 이스케이프 문자로 바꿈. 다음은 URL에서 안전하지 않은 문자를 입력했을 때, 인코딩된 예시이다.
 - 빈 문자열 : %20
 - ~ 기호 : %7
 - 위키 : %EC%9C%84%ED%82%A4

웹 개발 시, URI를 인코딩하고 디코딩하는 작업이 빈번하게 일어난다.
